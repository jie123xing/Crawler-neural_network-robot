```python
import torch
from torch.utils.data import Dataset
import pandas as pd
from torch.utils.data import DataLoader
import numpy as np
```

```python
# 定义自己的类
class MyDataset(Dataset):
    
    # 初始化
    def __init__(self, file_name):
        # 读入数据
        self.data = pd.read_csv(file_name)
    
    # 返回df的长度
    def __len__(self):
        return len(self.data)
    
    # 获取第idx+1列的数据
    def __getitem__(self, idx):
        return self.data.values[idx,1:10]

# 通过实例化对象来访问该类
# 假设同目录下存在名为median_benchmark.csv的文件
ds = MyDataset('originalweather.csv')
```

```python
ds=np.asarray(ds[:], dtype = np.float32)#####源文件float32 ,不知道为什么读取变为float64，这里再进行处理
```

```python
dl = DataLoader(ds, batch_size=10, shuffle=False)
```

```python
for i, data in enumerate(dl):
    #print(i, data)
    x=data[:,0:8]
    y=data[:,8]
    print(x,y)
    if i ==10:
        break
```
